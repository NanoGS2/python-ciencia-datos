```python
import matplotlib.pyplot as plt
from pandas._libs.algos import diff_2d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N ,[0.4, 0.5, 0.1])  # Por ejemplo, 40% masculino, 50% femenino, 10% no especificado)
    # Generar edades con distribución normal
    edad =np.random.normal(44, 6, N)
    edad=np.clip(np.around(edad),28,60)
    # Generar alturas, pesos y otras características
    altura = np.random.normal(180, 10, N)
    altura = np.clip(np.around(altura, decimals=1), 150.0, 210.0)

    peso = np.random.normal(80, 15, N)
    peso = np.clip(np.around(peso, decimals=1), 40.0, 130.0)

    horas_trabajo_estudio = np.random.normal(10, 2, N)
    horas_trabajo_estudio = np.clip(np.around(horas_trabajo_estudio, decimals=1), 4.0, 12.0)

    horas_ejercicio = np.random.normal(1, 0.5, N)
    horas_ejercicio = np.clip(np.around(horas_ejercicio, decimals=1), 0, 2)

    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    #La línea de arriba me tuvo hora y media dando vueltas
    return [sexo,edad,altura,peso,horas_descanso,horas_trabajo_estudio,horas_ejercicio]
```

```python
# Solicitar al usuario la cantidad N de datos a cargar
N = int(input("Ingrese el número de datos que desea generar "))

# Generar datos aleatorios para N personas con distribución normal y límites específicos
datos = generar_datos(N)
labels = ["Sexo","Edad","Altura","Peso","Horas_descanso","Horas_trabajo_estudio","Horas_ejercicio"]
dft = pd.DataFrame(datos,labels)
df = dft.transpose()


age_mean = df["Edad"].mean()
weight_mean = df["Peso"].mean()
height_mean = df["Altura"].mean()
w_hour_mean = df["Horas_trabajo_estudio"].mean()
ex_hour_mean = df["Horas_ejercicio"].mean()
rest_hour_mean = df["Horas_descanso"].mean()

print(" Las medias de las alturas, pesos , \n horas de descanso, horas trabajo y \n horas de ejercicios respectivamente, son : \n", height_mean , weight_mean , w_hour_mean,  ex_hour_mean , rest_hour_mean)


weigth_std = round(df["Peso"].std(),1)
heigth_std = round(df["Altura"].std(),1)


plt.hist(df["Peso"], bins = 50)
plt.xlabel("Frecuencia")
plt.title("Gráfico de frecuncias del peso")
plt.ylabel("Peso en kg")
plt.text(100,30,f"std: {weigth_std}kg")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def gradiente_descendente(w_, b_, alpha, x, y):
    '''Algoritmo del gradiente descendente para minimizar el error
       cuadrático medio'''
    N = x.shape[0]      # Cantidad de datos

    # Gradientes: derivadas de la función de error con respecto
    # a los parámetros "w" y "b"
    dw = -(2/N)*np.sum(x*(y-(w_*x+b_)))
    db = -(2/N)*np.sum(y-(w_*x+b_))

    # Actualizar los pesos usando la fórmula del gradiente descendente
    w = w_ - alpha*dw
    b = b_ - alpha*db

    return w, b

def calcular_modelo(w,b,x):
    '''Retorna el valor w*x+b correspondiente al modelo lineal'''
    return w*x+b


def calcular_error(y,y_):
    '''Calcula el error cuadrático medio entre el dato original (y)
       y el dato generado por el modelo (y_)'''
    N = y.shape[0]
    error = np.sum((y-y_)**2)/N
    return error




x = np.array(df["Edad"])
y = np.array(df["Altura"])




np.random.seed(2)
w = np.random.randn(1)[0]
b = np.random.randn(1)[0]

alpha = 0.0004
nits = 40000

error = np.zeros((nits,1))
for i in range(nits):
    # Actualizar valor de los pesos usando el gradiente descendente
    [w, b] = gradiente_descendente(w,b,alpha,x,y)

    # Calcular el valor de la predicción
    y_ = calcular_modelo(w,b,x)

    # Actualizar el valor del error
    error[i] = calcular_error(y,y_)


print("Epoch {}".format(i+1))
print("    w: {:.1f}".format(w), " b: {:.1f}".format(b))
print("    error: {}".format(error[i]))
print("=======================================")


x2 = np.arange(25,62,0.1)
y2 = x2*w + b


plt.scatter(df["Edad"],df["Altura"])
plt.title("Gráfico de Peso vs Altura")
plt.xlabel("Peso en kg")
plt.ylabel("Altura en cm")
plt.plot(x2,y2, c = "red")
plt.grid()
plt.show()

```

```python
#Usamos la funcion que nos dieron en clase para limpiar el DF

def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.05)
    q3 = data[column].quantile(0.95)
    iqr = q3 - q1
    lower_bound = q1
    upper_bound = q3
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Llamamos a la función para cada columna numérica donde desees detectar outliers
columns_to_check = ["Fare"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df_clean, col)
    outliers_list.append(outliers)

# Concatenamos todos los outliers encontrados en un solo DataFrame
outliers_df = pd.concat(outliers_list)

# Eliminamos los outliers del conjunto de datos original
data_cleaned = df.drop(outliers_df.index)

# Imprimimos la cantidad de outliers detectados
print("Número de outliers detectados:", len(outliers_df))

df_clean = data_cleaned





plt.figure(figsize=(8, 2))
sns.boxplot(x='Age', data=df_clean, palette='Set2')
plt.tight_layout()
plt.title("Age")
plt.show()
plt.figure(figsize=(8, 2))
sns.boxplot(x='Fare', data=df_clean, palette='Set2')
plt.tight_layout()
plt.title("Fare")
plt.show()
```
